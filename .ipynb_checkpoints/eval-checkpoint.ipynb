```python
# 1 load data
from utils.datareader import DataSet
from model.yolov3 import YoloJoint
from utils.fileio import json_save, json_load

annotation_file = "/Users/thisiszhou/Documents/dataset/" + \
                    "coco/annotations/instances_val2014.json"
photo_folder = "/Users/thisiszhou/Documents/dataset/coco/val2014"

#evaluate_model = "./coco83, nooc=1/yolov3-12000"
evaluate_model = "pre_train_weights/yolov3"
cocodata = DataSet(annotation_file, photo_folder)
#cocodata = DataSet(annotation_file, photo_folder, None, None)

#cocodata.save_categories_name()
cocodata.create_index()

# create model
yolo_model = YoloJoint(trainable=False)
yolo_model.build(image_size=416, coco_classes=80, joint_classes=0)
#yolo_model.build(image_size=416, coco_classes=80, joint_classes=1)
#yolo_model.init_sess("test_model_weghits/yolo")
#yolo_model.init_sess(pre_weights="./coco_weights/yolov3-2000")
#yolo_model.init_sess(pre_weights="./model_weights/yolov3-6000")
yolo_model.init_sess(evaluate_model)
#yolo_model.init_sess()
cate_dict = cocodata.coco_categories_dict
if_last = False
dt = dict()
dt["annotations"] = []
data_length = len(cocodata.img_ids)
num = 0
anno_id = 0
while not if_last:
    if num + 1 % 500 == 0:
        print("precessing: ", str(num), "/", str(data_length))
    imgs, scales, img_id, if_last = cocodata.get_coco_epoch(1)
    scale = scales[0]
    img_id = img_id[0]
    boxes, scores, first_cate = yolo_model.infer_for_eval(imgs)
    for i, box in enumerate(boxes):
        new_anno = dict()
        box = [box[0] * scale[0], box[1] * scale[1], 
               (box[2] - box[0]) * scale[0], (box[3] - box[1]) * scale[1]]
        new_anno['image_id'] = img_id
        new_anno['bbox'] = [float(x) for x in box]
        new_anno['category_id'] = cate_dict[first_cate[i] + 1]
        new_anno['score'] = float(scores[i])
        new_anno['area'] = float(box[2]*box[3])
        new_anno['id'] = anno_id
        dt["annotations"].append(new_anno)
        anno_id += 1
    if_last=True
json_save("data/eval.json", dt)
```
